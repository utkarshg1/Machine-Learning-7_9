### Seperate X values

```python
X = df[['Age', 'Income', 'Spend']]
```

```python
X
```

### Data Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Decide the number of clusters with silhoutee score

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
n = list(range(2, 15))
n
```

```python
scores = []
for i in n:
    model = AgglomerativeClustering(n_clusters=i)
    model.fit(X_pre)
    lb = model.labels_
    sil = silhouette_score(X_pre, lb)
    scores.append(sil)
```

```python
scores
```

### Plot silhoutee scores for different clusters

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(n, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee Score')
plt.title("Silhoutee score for different clusters")
plt.grid()
plt.show()
```

### Select 6 clusters as i get highest score of 0.42

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
model.labels_
```

### Save above results in dataframe

```python
X['Label'] = model.labels_
X.head()
```

```python
X[X['Label']==0]
```

```python
for i in range(6):
    print(f'Cluster Number : {i}\n')
    display(X[X['Label']==i])
    print('\n==============================================================\n')
```

### Centroids

```python
centroids = X.groupby(by='Label').mean()
centroids
```
